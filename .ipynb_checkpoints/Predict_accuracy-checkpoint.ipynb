```python
import numpy as np
from sklearn.linear_model import LinearRegression
X = np.array([])
y = np.array([])
cnt = 0
for record in open("record.txt","r").readlines()[1:]:
    X = np.append(X, np.asarray(record.split()[:-1]).astype(float),axis = 0)
    y = np.append(y,np.asarray(record.split()[-1:]).astype(float))
    cnt += 1
X = X.reshape(cnt,3)
reg = LinearRegression().fit(X, y)
reg.score(X, y)
```

```python
reg.predict(np.array([[1, 3, 0.75]]))
```

```python
accur_predict = {}
for i in range(10):
    for j in range(10):
        accu = reg.predict(np.array([[i,j,0.75]]))[0]
        if accu > 55:
            print(str(reg.predict(np.array([[i,j,0.75]]))[0]) + " " + str(i) + " " + str(j))
```
